```python
def extract(url):
    
    response = requests.get('%s' % url)
    response.encoding = 'utf-8' 
    data=response.text

    content = re.sub(html,'',data)
    content = content.replace('&nbsp;','')
    content = re.sub(r'&#\d+;', '', content)

    match_mda = re.findall(mda_regex, data, re.IGNORECASE | re.DOTALL | re.MULTILINE)
    if match_mda:
        mda = max(match_mda, key=len)
        mda = str(mda).replace('\n', '')
        mda = re.sub(r'(Item[0-9]\.|Item[0-9])', "", mda,)
    else:
        mda = None

    match_qqdmr = re.findall(qqd_regex, content, re.IGNORECASE | re.DOTALL | re.MULTILINE)

    if match_qqdmr:
        qqdmr = max(match_qqdmr, key=len)
        qqdmr = str(qqdmr).replace('\n','')
        qqdmr = re.sub(r'(Item[0-9]\.|Item[0-9])', "", qqdmr,)
    else:
        qqdmr = None

    match_rf = re.findall(riskfactor_regex, content, re.IGNORECASE | re.DOTALL | re.MULTILINE)


    if match_rf:
        rf = max(match_rf, key=len)
        rf = str(rf).replace('\n', '')
        rf = re.sub(r'(Item[0-9]\.|Item[0-9])', "", rf,)
    else:
        rf = None
    
    return(mda, qqdmr, rf)
print(url)

print(mda, qqdmr, rf)

```

```python
def positivescore(words):
    positive_count = 0
    for word in words:
        if word in positive:
            positive_count  += 1
    return(positive_count)
```

```python
def negativescore(words):
    negative_count = 0
    for word in words:
        if word in negative:
            negative_count  -= 1
    return(negative_count*(-1))
```

```python
def average_sentence_length(text):
        
    tokenizer = RegexpTokenizer(r'\w+')
    sentence_list = sent_tokenize(text)
    token_words = tokenizer.tokenize(text)
    cleaned_token_words = list(filter(lambda token: token not in stopwords, token_words))
    
    totalWordCount = len(cleaned_token_words)
    totalSentences = len(sentence_list)
    
    average_sent_length = 0
    if totalSentences != 0:
        average_sent_length = totalWordCount / totalSentences
        
    return round(average_sent_length)
```

```python
def complex_word(text):
    
    tokenizer = RegexpTokenizer(r'\w+')
    token_words = tokenizer.tokenize(text)
    cleaned_token_words = list(filter(lambda token: token not in stopwords, token_words))
    
    complexWord = 0
    complex_word_percentage = 0
    
    for word in cleaned_token_words:
        vowels=0
        if word.endswith(('es','ed')):
            pass
        else:
            for w in word:
                if(w=='a' or w=='e' or w=='i' or w=='o' or w=='u'):
                    vowels += 1
            if(vowels > 2):
                complexWord += 1
    if len(cleaned_token_words) != 0:
        complex_word_percentage = complexWord/complexWord
    
    return(complexWord, complex_word_percentage)
```

```python
def uncertaintyscore(words):
    uncertainty_score = 0
    for word in words:
        if word in uncertainty:
            uncertainty_score  += 1
    return(uncertainty_score)
```

```python
def constrainingscore(words):
    constraining_score = 0
    for word in words:
        if word in constraining:
            constraining_score  += 1
    return(constraining_score)
```

```python
def positive_word_proportion(positive_score,word_length):
    positive_word_proportion_score = 0
    if word_length !=0:
        positive_word_proportion = positive_score / word_length
        
    return positive_word_proportion
```

```python
def negative_word_proportion(negative_score,word_length):
    negative_word_proportion_score = 0
    if word_length !=0:
        negative_word_proportion_score = negative_score / word_length
        
    return negative_word_proportion_score
```

```python
def uncertainty_word_proportion(uncertainty_score,word_length):
    uncertainty_word_proportion_score = 0
    if word_length !=0:
        uncertainty_word_proportion_score = uncertainty_score / word_length
        
    return uncertainty_word_proportion_score
```

```python
def constraining_word_proportion(constraining_score,word_length):
    constraining_word_proportion_score = 0
    if word_length !=0:
        constraining_word_proportion_score = constraining_score / word_length
        
    return constraining_word_proportion_score
```

```python
def constraining_words_whole_report(words):
    constraining_words_whole_report_score = 0
    for word in words:
        if word in constraining:
            constraining_words_whole_report_score  += 1
    return(constraining_words_whole_report_score)
```

```python
def mda_fun(mda):
  
  if mda != None:
      #remove stopwords--------------
      mda = mda.lower()
      tokenizer = RegexpTokenizer(r'\w+')
      token_words=tokenizer.tokenize(mda)
      mda_words = list(filter(lambda token: token not in stopwords, token_words))

      #Positive score---------------
      mda_positive = positivescore(mda_words)

       #Negative score---------------
      mda_negative = negativescore(mda_words)

      #Polarity score---------------
      mda_polarity = (mda_positive - mda_negative) / ((mda_positive + mda_negative ) + 0.000001)

          #Subjectivity Score-----------
      mda_subjectivity = (mda_positive  + mda_negative)/ ((len(mda_words)) + 0.000001)


          #Average Sentence Length------
      mda_average_sentence_length = average_sentence_length(mda)

          #Complex words and Percentage of complex words---
      mda_complexWord, mda_percentage_complex_word = complex_word(mda)

          #Fog index
      #0.4 * (Average Sentence Length + Percentage of Complex words)
      mda_fogindex = 0.4 * (mda_average_sentence_length + mda_percentage_complex_word)

          #Word count---------
      mda_token_words = len(mda_words)

          #Uncertainty
      mda_uncertainty = uncertaintyscore(mda_words)

          #Constraining
      mda_constraining = constrainingscore(mda_words)

          #Positive_word_proportion
      mda_positive_word_proportion = positive_word_proportion(mda_positive,mda_token_words)

          #Negative_word_proportion
      mda_negative_word_proportion = negative_word_proportion(mda_negative,mda_token_words)


          #Uncertainty_word_proportion
      mda_uncertainty_word_proportion = uncertainty_word_proportion(mda_uncertainty,mda_token_words)

          #Constraining_word_proportion
      mda_constraining_word_proportion = constraining_word_proportion(mda_constraining,mda_token_words)

      return(mda_words, mda_positive, mda_negative, mda_polarity, mda_subjectivity, mda_average_sentence_length, mda_complexWord, mda_percentage_complex_word, mda_fogindex, mda_token_words, mda_uncertainty, mda_constraining, mda_positive_word_proportion, mda_negative_word_proportion, mda_uncertainty_word_proportion, mda_constraining_word_proportion)

  else:
    mda_words = []
    mda_positive = mda_negative = mda_polarity = mda_subjectivity = mda_average_sentence_length = mda_complexWord = mda_percentage_complex_word = mda_fogindex = mda_token_words = mda_uncertainty = mda_constraining = mda_positive_word_proportion = mda_negative_word_proportion = mda_uncertainty_word_proportion = mda_constraining_word_proportion = 0
    return(mda_words, mda_positive, mda_negative, mda_polarity, mda_subjectivity, mda_average_sentence_length, mda_complexWord, mda_percentage_complex_word, mda_fogindex, mda_token_words, mda_uncertainty, mda_constraining, mda_positive_word_proportion, mda_negative_word_proportion, mda_uncertainty_word_proportion, mda_constraining_word_proportion)
```

```python
def rf_fun(rf):
  
  if rf != None:
    
    #remove stopwords--------------
    tokenizer = RegexpTokenizer(r'\w+')
    rf = rf.lower()
    token_words=tokenizer.tokenize(rf)
    rf_words = list(filter(lambda token: token not in stopwords, token_words))

    #Positive score---------------
    rf_positive = positivescore(rf_words)
    
    #Negative score---------------
    rf_negative = negativescore(rf_words)

    #Polarity score---------------
    rf_polarity = (rf_positive - rf_negative)/ ((rf_positive + rf_negative ) + 0.000001)
    
    #Subjectivity Score-----------
    rf_subjectivity = (rf_positive  + rf_negative)/ ((len(rf_words)) + 0.000001)
    
    #Average Sentence Length------
    rf_average_sentence_length = average_sentence_length(rf)
    
    #Complex words and Percentage of complex words---
    rf_complexWord, rf_percentage_complex_word = complex_word(rf)
    

    #Fog index
    #0.4 * (Average Sentence Length + Percentage of Complex words)
    rf_fogindex = 0.4 * (rf_average_sentence_length + rf_percentage_complex_word)
    
    #Word count---------
    rf_token_words = len(rf)
    
    #Uncertainty
    rf_uncertainty = uncertaintyscore(rf_words)
    
    #Constraining
    rf_constraining = constrainingscore(rf_words)
    
    #Positive_word_proportion
    rf_positive_word_proportion = positive_word_proportion(rf_positive,rf_token_words)
    
    #Negative_word_proportion
    rf_negative_word_proportion = negative_word_proportion(rf_negative,rf_token_words)
    
    #Uncertainty_word_proportion
    rf_uncertainty_word_proportion = uncertainty_word_proportion(rf_uncertainty,rf_token_words)
    
    #Constraining_word_proportion
    rf_constraining_word_proportion = constraining_word_proportion(rf_constraining,rf_token_words)
    
    return(rf_words, rf_positive, rf_negative, rf_polarity, rf_subjectivity, rf_average_sentence_length, rf_complexWord, rf_percentage_complex_word, rf_fogindex, rf_token_words, rf_uncertainty, rf_constraining, rf_positive_word_proportion, rf_negative_word_proportion, rf_uncertainty_word_proportion, rf_constraining_word_proportion)

  else:
    rf_words =[]
    rf_positive = rf_negative = rf_polarity = rf_subjectivity = rf_average_sentence_length = rf_complexWord = rf_percentage_complex_word = rf_fogindex = rf_token_words = rf_uncertainty = rf_constraining = rf_positive_word_proportion = rf_negative_word_proportion = rf_uncertainty_word_proportion = rf_constraining_word_proportion = 0
    return(rf_words, rf_positive, rf_negative, rf_polarity, rf_subjectivity, rf_average_sentence_length, rf_complexWord, rf_percentage_complex_word, rf_fogindex, rf_token_words, rf_uncertainty, rf_constraining, rf_positive_word_proportion, rf_negative_word_proportion, rf_uncertainty_word_proportion, rf_constraining_word_proportion)
```
